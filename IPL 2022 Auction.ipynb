```python
def convertCurrancy(value):
    rupees = float(value.split(" ")[0])
    notation = value.split(" ")[1]
    if(notation == "L"):
        return rupees*100000
    else:
        return rupees*10000000
```

```python
data["Base Price"] = data["Base Price"].apply(lambda value:convertCurrancy(value)) 
data["Sold Price"] = data["Sold Price"].apply(lambda value:convertCurrancy(value)) 
```

```python
newdatas =[]
for i in range(len(data)):
    newdatas.append(data.iloc[i].to_dict())
len(newdatas)
```

```python
def convertInt(value):
    return int(value)
```

```python
# updating Sold Price of newdatas
for new in newdatas:
    new["Sold Price"] = convertInt(new["Sold Price"])
```

```python
soldPriceArray = []
for player in newdatas:
    soldPriceArray.append(player["Sold Price"])
```

```python
soldPriceArray
```

```python
soldPriceArray.sort(reverse=True)
```

```python
soldPriceArray
```

```python
def findData(price):
    for newdata in newdatas:
        if(newdata["Sold Price"] == price):
            return newdata
```

```python
findPlayer = []
for prices in soldPriceArray:
    findPlayer.append(findData(prices))
```

```python
for play in findPlayer:
    print("NAME:",play["Player"],' COUNTRY:' ,play["Country"], "Sold:", play["Sold Price"],  "BASE:", play["Base Price"])
```
